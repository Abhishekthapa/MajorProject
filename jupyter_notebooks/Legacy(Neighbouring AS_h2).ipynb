```python
import matplotlib.pyplot as plt

import numpy as np

plt.rcParams['axes.labelsize'] = 20
plt.rcParams['font.weight']='bold'
plt.rcParams['font.size']=14
plt.rcParams['legend.fontsize'] = 10
```

We look the performance between AS65000 and AS65002

```python
#between 65000 and 65002(h2 and h3)
#mdev of RTT for 5,10,20,30,50,100 packets

fig, ax = plt.subplots()


rtt_mdev=[0.046,0.044,0.032,0.064,0.057,0.054]
packets=[5,10,20,30,50,100]
l1, =plt.plot(packets,rtt_mdev)


rtt_mdev=[0.030,0.033,0.147,0.084,0.072,0.078]
packets=[5,10,20,30,50,100]
l2, =plt.plot(packets,rtt_mdev)
plt.legend((l1, l2), ('h2 to h3', 'h3 to h2'), loc='upper right')
plt.xlabel('No. of packets')
plt.ylabel('Standard Deviation of RTT')
plt.title('RTT Variance ')
```

```python
#between 65000 and 65001(h2 and h3)
#min and max RTT

barWidth = 0.25
 
# set height of bar
bars1 = [0.169,0.167,0.159,0.060,0.133,0.126]
bars2 = [0.292,0.299,0.263,0.345,0.364,0.376]

 
# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]



# Make the plot
plt.bar(r1, bars1, color='#7f6d5f', width=barWidth, edgecolor='white', label='min. RTT')
plt.bar(r2, bars2, color='#2d7f5e', width=barWidth, edgecolor='white', label='max. RTT')

# Add xticks on the middle of the group bars
plt.xlabel('No. of packets', fontweight='bold')
plt.ylabel('Time taken in ms', fontweight='bold')
plt.xticks([r + barWidth/2 for r in range(len(bars1))], ['5', '10', '20', '30', '50','100'])
 
#plt.savefig('minmaxRTT.png')
plt.title('Min. and Max. RTT (h2 to h3)')
plt.legend(loc='upper left')
plt.show()
```
