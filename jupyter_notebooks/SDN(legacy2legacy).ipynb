```python
import matplotlib.pyplot as plt

import numpy as np

plt.rcParams['axes.labelsize'] = 20
plt.rcParams['font.weight']='bold'
plt.rcParams['font.size']=14
plt.rcParams['legend.fontsize'] = 10
```

```python
#mdev of RTT for 5,10,20,30,50,100 packets
#SDN network as transit(h1 and h2)

fig, ax = plt.subplots()


rtt_mdev=[0.128,0.010,0.018,0.084,0.027,0.027]
packets=[5,10,20,30,50,100]
l1, =plt.plot(packets,rtt_mdev)


rtt_mdev=[0.138,0.014,0.014,0.021,0.014,0.020]
packets=[5,10,20,30,50,100]
l2, =plt.plot(packets,rtt_mdev)
plt.legend((l1, l2), ('h1 to h2', 'h2 to h1'), loc='upper right')
plt.xlabel('No. of packets')
# plt.ylabel('Standard Deviation of RTT')
plt.title('RTT Variance')
```

```python
barWidth = 0.25
 
# set height of bar
bars1 = [0.079,0.068,0.069,0.065,0.071,0.061]
bars2 = [0.405,0.100,0.148,0.540,0.249,0.232]

 
# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]


# Make the plot
plt.bar(r1, bars1, color='#7f6d5f', width=barWidth, edgecolor='white', label='min. RTT')
plt.bar(r2, bars2, color='#2d7f5e', width=barWidth, edgecolor='white', label='max. RTT')

# Add xticks on the middle of the group bars
plt.xlabel('No. of packets')
plt.ylabel('Time taken in ms')
plt.xticks([r + barWidth/2 for r in range(len(bars1))], ['5', '10', '20', '30', '50','100'])
 
#plt.savefig('PTR_transit.png')
plt.title('Min. and Max. RTT (h1 to h2)')
plt.legend(loc='upper left')
plt.show()
```
